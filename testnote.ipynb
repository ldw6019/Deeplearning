```python
import torch
import torch.nn as nn
import torch.optim as optim

# 1️⃣ 데이터 (y = 2x)
x = torch.tensor([[1.0], [2.0], [3.0]])
y = torch.tensor([[2.0], [4.0], [6.0]])

# 2️⃣ 모델 (선형 모델: y = wx + b)
model = nn.Linear(1, 1)

# 3️⃣ loss 함수
criterion = nn.MSELoss()

# 4️⃣ 옵티마이저
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
# =========================
# 🔥 학습 루프 (epoch)
# =========================
for epoch in range(100):  # 100번 반복
    # 1️⃣ forward
    output = model(x)
    loss = criterion(output, y)

    # 2️⃣ backward
    optimizer.zero_grad()
    loss.backward()

    # 3️⃣ step
    optimizer.step()

    # 4️⃣ 로그 출력
    if epoch % 10 == 0:
        print(f"epoch {epoch}, loss: {loss.item()}")

# 결과 확인
print("\n최종 weight:", model.weight.item())
print("최종 bias:", model.bias.item())
```
