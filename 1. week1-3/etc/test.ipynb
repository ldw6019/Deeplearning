```python
import torch
import pandas as pd
import numpy as np

# 1. 버전 확인
print(f"1. PyTorch 버전: {torch.__version__}")

# 2. 텐서 생성 테스트 (딥러닝의 기본 단위)
# 2x3 크기의 랜덤 숫자가 담긴 행렬을 만듭니다.
x = torch.rand(2, 3)
print("\n2. 생성된 테스트 텐서:\n", x)

# 3. 간단한 행렬 연산
y = x + 10
print("\n3. 연산 결과 (모든 요소에 10 더하기):\n", y)

# 4. 데이터 분석 라이브러리 확인
df = pd.DataFrame([[1, 2], [3, 4]], columns=['A', 'B'])
print("\n4. Pandas 데이터프레임 생성 성공:\n", df)

print("\n✨ 모든 환경이 완벽하게 준비되었습니다! 열공하세요! ✨")
```

```python
import torch

# 1. 0으로 가득 찬 텐서 만들기 (3행 3열)
zeros = torch.zeros(3, 3)

# 2. 랜덤한 숫자로 가득 찬 텐서 만들기
random_box = torch.rand(3, 3)

# 3. 두 텐서 더하기
result = zeros + random_box

print("--- 딥러닝의 기본 단위 '텐서' 테스트 ---")
print(result)
print("\n축하합니다! 이제 딥러닝 수식을 코드로 구현할 준비가 되었습니다.")
```
