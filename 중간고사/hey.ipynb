## **[Part 2] 서술형 (총 3문항, 각 10점 = 30점)**
> **왜 그런지 기술적인 근거를 포함하여 명확하게 서술하세요.**

1. **[구조 관리]** 신경망 내부에 여러 개의 은닉층 레이어를 동적으로 생성하여 리스트 형태로 관리하려고 합니다. 이때 일반적인 Python `list`를 사용하면 안 되고 반드시 PyTorch의 `nn.ModuleList`를 사용해야 하는 이유를 **'파라미터 추적 및 업데이트'** 관점에서 서술하세요.
   1. torch는 list의 내부 연산을 제대로 추적할수 없음. 모델에게 parameters를 보내려 해도 아무 일이 일어나지 않고, 정상적인 학습이 불가능함. nn.modulelist는 제대로 진행이 추적돼 파라미터들을 관리해 loss, 역전파 등이 잘 작동함
2. **[정규화 원리]** 과적합을 방지하기 위해 옵티마이저에 `weight_decay`를 설정하여 **L2 정규화(Ridge)**를 적용했습니다. 이 기법이 수학적으로 손실 함수에 무엇을 더하는지 설명하고, 이것이 어떻게 과적합을 억제하는지 서술하세요.
   1. L2 정규화는 weight_decay를 통해 큰 가중치에 벌점을 줘 가중치를 0에 가깝게 유지.
3. **[모드 전환]** 훈련(Train) 루프에서는 `model.train()`을, 평가(Validation/Test) 루프에서는 `model.eval()`을 반드시 호출해야 합니다. **드롭아웃(Dropout)**의 동작 방식을 예로 들어, 왜 이 두 모드를 다르게 설정해야 하는지 서술하세요.
   1. 학습 루프 중에는 과적합을 막기 위해 드랍아웃으로 확률적 비활성화를 시켜줬지만, 평가 단계에서는 모든 뉴런이 사용되어야 하는데 무작위로 비활성화될시 매 시도마다 균등한 결과를 얻을수 없다.
